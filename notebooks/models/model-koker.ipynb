# Teddy Koker

Teddy Koker [1] did implement an simple reinforcement learning algorithm on BTCUSD and made a nice article about it on his blog [2]. 

Here, we reproduce his experiment on different data. His code is located in the `src.tars.models` directory. The data is from Bitcoincharts [3].

- [1] https://teddykoker.com
- [2] https://teddykoker.com/2019/06/trading-with-reinforcement-learning-in-python-part-ii-application/
- [3] https://api.bitcoincharts.com/v1/csv/

## Imports

```python
import pandas as pd
from src import models
```

## Main  

Load the data

```python
data = pd.read_csv("../data/raw/bitstampUSD.csv", names=["utc", "price", "volume"]).set_index('utc')
```

Load the model

```python
model = models.koker.Koker()
model.load(data)
```

Train the model

```python
model.train(model.x_train)
```

Check the results

```python
model.plot(model.x_train, model.theta)
```

```python
model.plot(model.x_test, model.theta)
```
