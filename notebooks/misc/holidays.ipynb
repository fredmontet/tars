```python
from darts.utils import timeseries_generation
import matplotlib.pyplot as plt
import datetime
import pandas as pd
import seaborn as sns
```

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 12, 31)
index = pd.date_range(start, end)
```

```python
ch = timeseries_generation.holidays_timeseries(index,country_code='CH')
de = timeseries_generation.holidays_timeseries(index,country_code='DE')
at = timeseries_generation.holidays_timeseries(index,country_code='AT')
fr = timeseries_generation.holidays_timeseries(index,country_code='FR')
be = timeseries_generation.holidays_timeseries(index,country_code='BE')
lu = timeseries_generation.holidays_timeseries(index,country_code='LU')
it = timeseries_generation.holidays_timeseries(index,country_code='IT')
```

```python
holidays = ch.stack(de).stack(at).stack(fr).stack(be).stack(lu).stack(it)
df = holidays.pd_dataframe()
df.columns = ['CH', 'DE', 'AT', 'FR', 'BE', 'LU', 'IT']
```

```python
df.groupby(pd.Grouper(freq='M')).sum()
```

```python
df.groupby(pd.Grouper(freq='M')).sum().plot(figsize=(15,7));
```
